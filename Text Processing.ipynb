### What if we try to compare genres? 


```python
from nltk import FreqDist
modals = ["may", "could", "will"]
genres = ["adventure", "news", "romance"]

for g in genres:
    words = brown.words(categories=g)
    frequencies = FreqDist([w.lower() for w in words if w.lower() in modals])
    print(g, frequencies.most_common(3))
    
```

## Bigrams

```python
from nltk.corpus import stopwords 
from nltk.collocations import BigramCollocationFinder, BigramAssocMeasures
ignored_words = stopwords.words('english')
finder = BigramCollocationFinder.from_words(emma_words, 4) #last param - windows size
finder.apply_freq_filter(2) 
finder.apply_word_filter(lambda w: len(w) < 3 or w.lower() in ignored_words)

# nbest = top n ngrams scored by the given function
bigram_measures = BigramAssocMeasures()
collocations = finder.nbest(bigram_measures.likelihood_ratio, 60) 
colloc_strings = [w1+' '+w2 for w1, w2 in collocations]
print(colloc_strings)
```

### And now trigrams. Without Personal Pronouns

```python
from nltk.corpus import stopwords 
from nltk.collocations import TrigramCollocationFinder, TrigramAssocMeasures
ignored_words = stopwords.words('english')
finder = TrigramCollocationFinder.from_words(emma_words, 3) #last param - windows size
finder.apply_freq_filter(2) 
finder.apply_word_filter(lambda w: len(w) < 4 or w.lower() in ignored_words or w.lower() != w)

# nbest = top n ngrams scored by the given function
bigram_measures = TrigramAssocMeasures()
collocations = finder.nbest(bigram_measures.likelihood_ratio, 60) 
colloc_strings = [w1+' '+w2+' ' + w3 for w1, w2, w3 in collocations]
print(colloc_strings)
```
